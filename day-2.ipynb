```python
string = input("Enter a string: ")
freq = {}
for ch in string:
    freq[ch] = freq.get(ch, 0) + 1
for ch in string:
    if freq[ch] == 1:
        print("First non-repeating character:", ch)
        break
else:
    print("No non-repeating character found")
```

```python
acbal = 10000

def deposit():
    global acbal
    amt = int(input("Enter your deposit amount: "))

    if amt % 100 != 0:
        print("Please enter multiples of 100 only")
    elif amt > 50000:
        print("Maximum deposit limit is 50000")
    else:
        acbal += amt
        print("Deposit successful")

    print("Available balance:", acbal)


def withdraw():
    global acbal
    amt = int(input("Enter withdraw amount: "))

    if amt % 100 != 0:
        print("Withdraw amount must be multiples of 100")
    elif amt > 20000:
        print("Withdraw limit per transaction is 20000")
    elif amt > acbal:
        print("Insufficient balance")
    else:
        acbal -= amt
        print("Withdrawal successful")

    print("Available balance:", acbal)


def balance_enquiry():
    print("Available balance:", acbal)


def viewOptions():
    while True:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Balance Enquiry")
        print("0. EXIT")

        option = int(input("Choose your option: "))

        if option == 1:
            deposit()
        elif option == 2:
            withdraw()
        elif option == 3:
            balance_enquiry()
        elif option == 0:
            print("Thank you, visit again")
            break
        else:
            print("Invalid option")


viewOptions()
```
